### Creates preprocessing pipeline, impute most frequent values and if any unknown values then -1

```python
ordinal_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('ord', OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1))
])
```

### Fills missing values with frequent and ignore unknown categories

```python
ohe_pipeline = Pipeline(steps=[
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('one-hot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])
```

```python
col_trans = ColumnTransformer(transformers=[
    ('impute', SI, ['Age']), # Fill missing values in the 'Age' column using the imputer (SI)
    ('ord_pipeline', ordinal_pipeline, ode_cols),  # Apply ordinal encoding to specified ordinal columns (ode_cols)
    ('ohe_pipeline', ohe_pipeline, ohe_cols),  # Apply ordinal encoding to specified ordinal columns (ode_cols)
   
      # Keep these columns unchanged (pass them through as they are)
     ('passthrough', 'passthrough', ['Pclass', 'Cabin_Assigned', 'Name_Size', 'Age', 'Fare', 'TicketNumberCounts'])
    ],
    remainder='drop', # Drop any columns not specified in the transformations
    n_jobs=-1)  # Use all available CPU cores to speed up processing
```

```python
param_grid = {
    'n_estimators': [150, 200, 300, 500], # Number of trees
    'min_samples_split': [5, 10, 15], # Minimujm samples required to split a node. Higher values prevent deep splits and reduce overfitting
    'max_depth': [10, 13, 15, 17, 20], # Maximum depth of each tree. Controls how deep trees can grow (prevents overfitting if limited).
    'min_samples_leaf': [2, 4, 5, 6], # Minimum samples needed to be in a leaf node. Ensures leaf nodes are not too small, improving generalization.
    'criterion': ['gini', 'entropy'], # Gini tells us how mixed the data is in a node. Entropy measures how much uncertainty is in a group.
}
```
